```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/simra/Downloads/Expanded_data_with_more_features.csv/Expanded_data_with_more_features.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# drop - delete column
df = df.drop("Unnamed: 0", axis =1)
df.head()
```

```python
# column transformation
# change weekly study hours column
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct", "05-10")
df.head()
```

```python
# gender distribution
plt.figure(figsize= (6,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# from the above chart we have analysed that:
# the number of females in the data is more than the number of  males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore": 'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb,annot = True )
plt.title("Realtion between Parent's Eduction and Student Score")
plt.show()
```

```python
# from the above chart we have concluded that the eduction of the parents 
# have a good impact score
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore": 'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb1,annot = True )
plt.title("Realtion between Parent's Marital Status and Student Score")
plt.show()
```

```python
# from the above chart we can conclude that there is no/niglegible impact on the student score 
# by the parent's marital status
```

```python
# Distribution of Ethnic Groups
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()
# groupA['EthnicGroup']

l = ["group A","group B","group C","group D","group E"]
mlist = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]

print(mlist)
plt.pie(mlist,labels = l, autopct = "%1.2f%%")
plt.title("Distribution of ethnic Groups")
plt.show()
# groupA
```
